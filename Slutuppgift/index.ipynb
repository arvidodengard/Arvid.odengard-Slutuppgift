```python
# Importerar de biblotek som används.
import pandas as pd 
from plotly.subplots import make_subplots
import plotly.graph_objects as go

pd = pd.read_csv

# Gör en funktion för att läsa in filer.
def läs_in_filer (filer):
    pd(filer)
    
avlidna_per_ålder = ("National_Total_Deaths_by_Age_Group.csv")
dagliga_fall_per_region = ("Regional_Daily_Cases.csv")
totala_fall_per_region = ("Regional_Totals_Data.csv")

läs_in_filer(avlidna_per_ålder)
läs_in_filer(dagliga_fall_per_region)
läs_in_filer(totala_fall_per_region)

# Döper om filnamnen en andra gång så att jag slipper att skriva med pd

avlidna = pd(avlidna_per_ålder)
regioner = pd(dagliga_fall_per_region)
totala_fall = pd(totala_fall_per_region)

# Ger namn till specefika listor

ålder = avlidna['Age_Group'] 
antal_avlidna = avlidna['Total_Deaths']

datum = regioner['Date']
fall = regioner['Sweden_Total_Daily_Cases']

Region = totala_fall['Region']
cases = totala_fall['Total_Cases']
avlidna = totala_fall['Total_Deaths']

# Gör 4 olika grafer

fig = make_subplots(
    rows=2, cols=2,
    specs=[[{"type": "Scatter"}, {"type": "Bar"}],
           [{"type": "xy"}, {"type": "Pie"}]], 
           subplot_titles=("Avlidna per ålder", "Fall per dag", 
            "Avlidna per region", "Totala fall per region"))

fig.add_trace(
    go.Scatter(x= ålder, y= antal_avlidna),
    row=1, col=1)

fig.add_trace(
    go.Bar(x= datum, y= fall),
    row=1, col=2)

fig.add_trace(
    go.Line(x= Region, y= avlidna),
    row=2, col=1)

fig.add_trace(
    go.Pie(labels = Region, values= cases),
    row=2, col=2)

fig.update_layout(height=750, width=800, title_text="Covid data")
fig.show()
```
